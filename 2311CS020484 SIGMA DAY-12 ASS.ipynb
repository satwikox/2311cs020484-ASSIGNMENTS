```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fp='C:/Users/ADMIN/Downloads/Day_11_banking_data.csv'
banking_data=pd.read_csv(fp)
```

```python
# Set Matplotlib styles
plt.style.use('ggplot')

# 1. Bar Plot: Total sum of Transaction_Amount per Account_Type
transaction_sum_per_account_type = banking_data.groupby('Account_Type')['Transaction_Amount'].sum()

# Create the bar plot
plt.figure(figsize=(8, 5))
transaction_sum_per_account_type.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Total Transaction Amount per Account Type', fontsize=16)
plt.ylabel('Total Transaction Amount', fontsize=12)
plt.xlabel('Account Type', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# 2. Pie Chart: Percentage of Transactions per Branch
transactions_per_branch = banking_data['Branch'].value_counts()

# Create the pie chart
plt.figure(figsize=(8, 8))
transactions_per_branch.plot(
    kind='pie', 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=plt.cm.Paired.colors,
    wedgeprops={'edgecolor': 'black'}
)
plt.title('Percentage of Transactions per Branch', fontsize=16)
plt.ylabel('')  # Remove default y-label
plt.tight_layout()
plt.show()
```
